```python
fact=1
num=int(input("Enter the number:"))
for i in range(1,num+1):
    fact=fact*i
print(fact)

rev_fact=str(fact)[::-1]
if str(fact)==rev_fact:
    print("palindrome")
else:
    print("no palindrome")
```
